### 1. 
Create a one-dimensional array with range 2 to 5  elements; Create a 2 by 4 two-dimensional array with range 2 to 9; Create a 4 by 3 by 2 three-dimensional array with range 0 to 23.

Print their properties: shape, dimension, size, and data type.

Solution:

```python
import numpy as np

print("Part 1")
a = np.arange(2,6) # one-dimensional array with range 2 to 5 elements
print(a)
print("shape :", a.shape) #shape
print("dimension:", a.ndim) #dimension
print("size:", a.size) #size
print("data-type:", a.dtype) #data type
print()

print("Part 2")
b = np.arange(2,10).reshape(2,4) # 2 by 4 two-dimensional array with range 2 to 9
print(b)
print("shape :", b.shape) #shape
print("dimension:", b.ndim) #dimension
print("size:", b.size) #size
print("data-type:", b.dtype) #data type
print()

print("Part 3")
c = np.arange(0,24).reshape(4,3,2) # 4 by 3 by 2 three-dimensional array with range 0 to 23
print(c)
print("shape :", c.shape) #shape
print("dimension:", c.ndim) #dimension
print("size:", c.size) #size
print("data-type:", c.dtype) #data type
```

### 3.
Create an array from 0 to 9 and extract the even elements for printing; then replace these elements by -1.
(Hint: use boolean index and element-wise comparison)

Create two arrays: x = [5,6,7,8,3,4] and y = [5,3,4,5,2,4], find the positions of the same elements in the two array.
(Hint: np.where() and element-wise comparison)

Create two arrays: [0, 10, 20, 40, 60] and [0, 40]. test whether each element of x is also present in y.
(Hint: np.isin() and use the two arrays as input)

Solution:

```python
print("Part 1")
a = np.arange(0,10)
print(a)

b = []
for i in a:
    if i % 2 == 0:
        b.append(i) # extract the even elements for printing
        a[i] = -1 # replace these elements by -1

b = np.array(b)        
print(b)    
print(a)
print()

print("Part 2")
x = np.array([5,6,7,8,3,4])
y = np.array([5,3,4,5,2,4])
print(np.where(x == y)) 
print()

print("Part 3")
x = np.array([0, 10, 20, 40, 60])
y = np.array([0, 40])
print(np.isin(x, y)) # test whether each element of x is also present in y
```

### 4.
Creat an one-dimensional array to hold 24 random integer numbers between 0 and 100, then do the following:

1. Find the position of minimum and maximum;
2. Reshape it into 6 by 4 array; 
3. Sort the array along axis 0;
4. Calculate mean, average, std of the array along axis = 0;
5. Reshape it into 4 by 3 by 2 and print the new view;
6. Transpose the array in step 5;
7. Save the file into a text output file "data.csv" with the delimiter = ','.
(Hint: x is a 3D array, you can not directly use `np.savetxt` to save it, you have to iterate over the last axis to slice the array and save.)

Solution:

```python
a = np.random.randint(0,100, 24)
print(a)
print()

print("Part 1")
# 1. Find the position of minimum and maximum
print("Postion of min at index =", a.argmin())
print("Postion of max at index =", a.argmax())
print()

print("Part 2")
# 2. Reshape it into 6 by 4 array;
b = a.reshape(6,4)
print(b)
print()

print("Part 3")
# 3. Sort the array along axis 0;
c = np.sort(b, axis=0)
print(c)
print()

print("Part 4")
# 4. Calculate mean, average, std of the array along axis = 0;
print("mean along axis=0 :", c.mean(axis=0))
print("average along axis=0 :", np.average(c, axis=0))
print("std along axis=0 :", c.std(axis=0))
print()

print("Part 5")
# 5. Reshape it into 4 by 3 by 2 and print the new view;
d = c.reshape(4,3,2)
print(d)
print()

print("Part 6")
# 6. Transpose the array in step 5;
e = np.transpose(d)
print(e)
print()

print("Part 7")
# 7. Save the file into a text output file "data.csv" with the delimiter = ','.
#np.savetxt("data.csv", d, delimiter=",")
with open('data.csv', 'w') as savefile:
        for z in range(4):
            np.savetxt(savefile, e[:, :, z], delimiter=",") # iterate over the last axis
            
print("file 'data.csv' created")
```

Create two arrays by the codes:

`array1 = np.arange(20).reshape(4,5)`

Test if any element in array1 and array 2 is evaluated to `True` by `np.any()`.

Then replace all the `np.nan` in array2 by the mean of array1. (Hint: use `np.isnan()` to test array2 first)



```python
print("Part 1")
array1 = np.arange(20).reshape(4,5)
print(array1)
print()

array2 = np.array([[1,2,np.nan],[4,5,6],[np.nan, 7, np.nan]])
print(array2)
print()

print("Part 2")
print("For array1", np.all(array1)) 
print("For array2", np.all(array2))
print()

print("Part 3")
print("For array1", np.any(array1)) 
print("For array2", np.any(array2)) 
print()

print("Part 4")
print("The mean of array1 =", array1.mean())
array2isnan = np.isnan(array2)
print(array2isnan)
print()

for i in range(3):
    for j in range(3):
        if array2isnan[i][j] == True:
            array2[i][j] = array1.mean()

print(array2)
```

a. Creat an array with 3 by 4 random numbers array within the range of (0, 1). Swap the three rows, that is, row index (0, 1, 2) -> row index (2, 0, 1). Swap first column and third column, that is, column index (0, 1, 2, 3) -> column index (2, 1, 0, 3)



```python
print("Part 1")
# Create an array with 3 by 4 random numbers array within the range of (0, 1)
a = np.random.rand(3,4).reshape(3,4)
print(a)
print()

print("Part 2")
# Swap the three rows, that is, row index (0, 1, 2) -> row index (2, 0, 1)
a[[0, 1, 2]] = a[[2, 0, 1]]
print(a)
print()

print("Part 3")
# Swap first column and third column, that is, column index (0, 1, 2, 3) -> column index (2, 1, 0, 3)
a[:, [0, 1, 2, 3]] = a[:, [2, 1, 0, 3]]
print(a)
print()
```

### 8.

Iris data is a prototype data set for machine learning beginners in building classfication models. It has five columns: sepallength, sepalwidth, petallength, petalwidth, and species. Data in the first four columns are float type data and the last column is string. Do the work below:

1. Use `genfromtxt()` to read data from the url `'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'` to a numpy array. 

2. Store the numeric values in one array and string values in the other array.

3. Calculate the mean, median, std of the first column, which is sepal length.

4. Normalize the data in the first columns, so that data are in the range of [0, 1]

5. the 5th and 95th percentile of iris's sepallength (the first column)


```python
print("Part 1")
iris = np.genfromtxt('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', 
                     delimiter=",", dtype = None, encoding = None)
print(iris)
# print(iris.shape) 150 rows
print()

print("Part 2")
a = iris.tolist() # convert to list
numerics = []
strings = []
for i in a:
    numerics.append(list(i[0:4]))
    strings.append(i[-1])
# Store the numeric values in one array and string values in the other array
num_ = np.array(numerics)
str_ = np.array(strings)
print(num_)
print(str_)
print()

print("Part 3")
mean = num_.mean(axis=0)
print('The mean of sepal length is', mean[0])
median = np.median(num_,axis=0)
print('The median of sepal length is', median[0])
std = num_.std(axis=0)
print('The std of sepal length is', std[0])
print()

# Part 4
print("Part 4")
max_val = num_.max(axis=0)
max_sepal_length = max_val[0]
min_val = num_.min(axis=0)
min_sepal_length = min_val[0]

normalized_num_ = np.copy(num_)
for i in range(150):
    x = normalized_num_[i][0]
    normalized_num_[i][0] = (x - min_sepal_length)/(max_sepal_length - min_sepal_length)
    
print(normalized_num_)
print()

# Part 5
N5_percentile = np.percentile(num_, 5, axis = 0)
print("The 5th percentile of sepal length is", N5_percentile[0])
N95_percentile = np.percentile(num_, 95, axis = 0)
print("The 95th percentile of sepal length is", N95_percentile[0])
```
