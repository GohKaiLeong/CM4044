## Matplotlib Exercise

For this exercise, we are using Matplotlib to visualize Company Sales Data, which is stored in a text file, "company_sales_data.csv". It is available in NTULearn.

The data file has 9 columns of data with a header for every column's name:

"month_number","facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer","total_units","total_profit"

### 1.
Read total profit of all months from "company_sales_data.csv" and show it using a line plot:

Total profit data provided for each month. Generated line plot must include the following properties: â€“

- X label name = Month
- Y label name = Total profit
- figure title: Company profit per month
- x ticks: month number
- y ticks: 100000, 200000, 300000, 400000, 500000

Solution:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline    
import pandas as pd

profits = pd.read_csv("company_sales_data.csv")
profits
```

```python
# plt method
plt.figure(figsize=(7,5),dpi=100)

X=profits['month_number']
Y=profits['total_profit']

plt.plot(X, Y)

plt.title('Company profit per month')
plt.xlabel('Month')
plt.ylabel('Total profit')

plt.xticks(X)
plt.yticks([100000, 200000, 300000, 400000, 500000])

plt.show()
```

```python
# OOP method
fig = plt.figure(figsize=(7,5), dpi=100)

X=profits['month_number']
Y=profits['total_profit']

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(X,Y)

ax.set_title('Company profit per month')
ax.set_xlabel('Month')
ax.set_ylabel('Total profit')

ax.set_xticks(X)
ax.set_yticks([100000, 200000, 300000, 400000, 500000])

plt.show()
```

### 2.

Generated line plot for company profit data to include following Style properties: 

- Line Style dotted and Line-color should be red
- Show legend at the lower right location. Legent text: Profit data of last year
- X label name = Month
- Y label name = Total profit
- Add a circle marker.
- Line marker color as read
- Line width should be 3

Solution:

```python
# plt method
plt.figure(figsize=(7,5),dpi=100)

X=profits['month_number']
Y=profits['total_profit']

plt.plot(X, Y, color='red', linewidth=3, linestyle=":") # Line Style dotted and Line-color should be red, line width = 3

plt.legend(['Profit data of last year'], loc="lower right")  
plt.xlabel('Month')
plt.ylabel('Total profit')

plt.scatter(X, Y, 50, color='red') # add red circle

plt.title('Company profit per month')
plt.xticks(X)
plt.yticks([100000, 200000, 300000, 400000, 500000])

plt.show()
```

```python
# OOP method
fig = plt.figure(figsize=(7,5), dpi=100)

X=profits['month_number']
Y=profits['total_profit']

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(X,Y,color ='red', linewidth=3, linestyle=":")

ax.legend(['Profit data of last year'], loc="lower right")  
ax.set_xlabel('Month')
ax.set_ylabel('Total profit')

ax.scatter(X, Y, 50, color='red')

ax.set_title('Company profit per month')
ax.set_xticks(X)
ax.set_yticks([100000, 200000, 300000, 400000, 500000])

plt.show()
```

### 5.

Read face cream and face wash product sales data and show it using the bar chart. Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

- legend text: "Face Cream Sales Data" and "Face Wash Sales Data"
- legend position well so that legend will not overlap with bars in the plot
- x label "Month"
- y label "Sales units in number"
- figure title "Facewash and FaceCream Sales Data"
- xticks: month
- yticks: use system default value

Solution:

```python
# plt method
plt.figure(figsize=(12,5),dpi=100)

X=profits['month_number']
Y1=profits['facecream']
Y2=profits['facewash']

plt.title('Facewash and FaceCream Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales units in number')

rects1 = plt.bar(X-0.21, Y1, 0.4, color='red', label = 'Face Cream Sales Data') 
rects2 = plt.bar(X+0.21, Y2, 0.4, color='orange', label = 'Face Wash Sales Data')

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 0),  
                    textcoords="offset points",
                    ha='center', va='bottom')
            
autolabel(rects1)
autolabel(rects2)

plt.legend(loc="best", fontsize=8)  

plt.xticks(X)

plt.show()
```

```python
# OOP method
fig = plt.figure(figsize=(12,5),dpi=100)
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

X=profits['month_number']
Y1=profits['facecream']
Y2=profits['facewash']

ax.set_title('Facewash and FaceCream Sales Data')
ax.set_xlabel('Month')
ax.set_ylabel('Sales units in number')

rects1 = ax.bar(X-0.21, Y1, 0.4, color='red', label = 'Face Cream Sales Data') 
rects2 = ax.bar(X+0.21, Y2, 0.4, color='orange', label = 'Face Wash Sales Data')

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 0),
                    textcoords="offset points",
                    ha='center', va='bottom')
            
autolabel(rects1)
autolabel(rects2)

ax.legend(loc="best", fontsize=8)  

ax.set_xticks(X)

plt.show()
```

### 7.

Use subplots to display sales data for Bathingsoap, Facewash, Facecream and Toothpaste.

Solution:

```python
fig, axes = plt.subplots(2,2, figsize=(8,8))
fig.tight_layout()
axes = axes.flatten()

x = profits["month_number"]

y0 = profits["bathingsoap"] 
axes[0].plot(x, y0, color ='r') 
axes[0].set_title('Bathing soap sales') 
axes[0].set_xticks(x)
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Sales units')
  
y1 = profits["facewash"] 
axes[1].plot(x, y1, color ='b') 
axes[1].set_title('Face wash sales') 
axes[1].set_xticks(x)
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Sales units')
  
y2 = profits["facecream"] 
axes[2].plot(x, y2, color ='g') 
axes[2].set_title('Face cream sales') 
axes[2].set_xticks(x)
axes[2].set_xlabel('Month')
axes[2].set_ylabel('Sales units')
  
y3 = profits["toothpaste"] 
axes[3].plot(x, y3, color ='k') 
axes[3].set_title('Toothpaste sales')
axes[3].set_xticks(x)
axes[3].set_xlabel('Month')
axes[3].set_ylabel('Sales units')
  
fig.subplots_adjust(hspace=.5,wspace=0.5) 

plt.show()
```

### 8.

Read all product sales data and show them using the stack plot.

- legend is on best position, so that it doesn't overlap with plot
- different colours are used for different product
- Figure Title: All product sales data in stack plot
- x title: Month
- y title: Sales units in number

Solution:

```python
fig = plt.figure(figsize=(7,5),dpi=100)
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

x = profits["month_number"]

y0 = profits["facecream"] 
y1 = profits["facewash"] 
y2 = profits["toothpaste"] 
y3 = profits["bathingsoap"] 
y4 = profits["shampoo"]
y5 = profits["moisturizer"]

legends = ["Face cream", "Face wash", "Toothpaste", "Bathing soap", "Shampoo", "Moisturizer"]
ax.stackplot(x, y0, y1, y2, y3, y4, y5, labels=legends)
ax.set_title("All product sales data in stack plot")
ax.set_xlabel("Month")
ax.set_ylabel("Sales units in number")
ax.set_xticks(x)
ax.legend(loc = "upper left", fontsize=8)

plt.show()
```
